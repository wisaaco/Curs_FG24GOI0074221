```python
import pandas as pd
```

```python
my_dict={'sale_id':[1,2,3,4,5,6,7,8,9],
         'c_id':[2,2,1,4,2,3,2,3,2],
         'p_id':[3,4,3,2,3,3,2,2,3],
	 'product':['Monitor','CPU','Monitor','RAM','Monitor','Monitor','RAM','RAM','Monitor'],
         'qty':[2,1,3,2,3,2,3,2,2],
         'store':['ABC','DEF','ABC','DEF','ABC','DEF','ABC','DEF','ABC']}
sales = pd.DataFrame(data=my_dict)

my_dict={'p_id':[1,2,3,4,5,6,7,8],
  'product':['Hard Disk','RAM','Monitor','CPU','Keyboard','Mouse','Motherboard','Power supply'],
  'price':[80,90,75,55,20,10,50,20]}
product = pd.DataFrame(data=my_dict)

my_dict={'c_id':[1,2,3,4,5,6,7,8],
         'Customer':['Rabi','Raju','Alex','Rani','King','Ronn','Jem','Tom']}
customer = pd.DataFrame(data=my_dict)
```

```python
# 1. Explora el dataframe en el que podem esbrinar les quantitats venudes de cada producte.
sales.head(9)
```

```python
# 2. Llista de quantitats i vendes totals de cada producte

llista_productes = sales['product'].unique().tolist()

print(f"Tenim els següents productes: {llista_productes}")

sales.groupby(['product'])[['qty']].sum()
```

```python
#3. Llista de quantitats venudes (nombre) i de doblers obtinguts de ventes de cada producte.
ventes = sales.groupby(['product','p_id', 'store'])[['qty']].sum()

#print(ventes)
unio = pd.merge(ventes, product, how='left',on='p_id')
#print(unio)

unio['total_sale']= unio['qty'] * unio['price']
res = unio.groupby("product").sum()
res[["qty", "total_sale"]]
res
```

```python
#veure l'efecte reset_index
my_sale=sales.groupby(['product','p_id', 'store'])[['qty']].sum().reset_index()
my_sale
```

```python
#4. Llista de quantitats venudes per producte i per botiga.

res = sales.groupby(['product','store'])[['qty']].sum()
res
```

```python
# 5. Llista dels productes que no es venen.

dades = sales.merge(product, on='p_id', how='right')
#print(dades)
#print(dades['sale_id'].isna())
dades2 =dades[dades['sale_id'].isna()] # productes no venuts
#print(dades2)
res = dades2.loc[:,'product_y'] # només mostram els productes
#print(f"Els productes que no es venen son {res.tolist()}")

for producte in res.tolist():

  print(f"Aquest producte no s'ha venut: {producte}")
```

```python
# 6. Llistat de clients que no han comprat cap producte.

dades = sales.merge(customer, on='c_id',how='right')
print(dades)
clients=dades[dades['sale_id'].isnull()] # clients
print(clients)
print(clients.loc[:,'Customer'].tolist()) # to display customers who has not purchased
```

```python
#EXTRA
my_sum=pd.merge(sales,product,how='left',on=['p_id'])
my_sum['sales_total']=my_sum['qty']*my_sum['price']
print(my_sum.groupby(['store'])[['qty','sales_total']].sum())
```
