```python
#1 Seleccioneu la columna “species” del dataframe “penguins”.
# Quantes espècies diferents de pingüins hi ha al conjunt de dades?
penguins.species.unique()


penguins.groupby("island").aggregate(
    {               # Diccionari
    "island":"count"     # Columna : Operacion de agregació
    }
)
```

```python
#2 Selecciona les columnes “species”, “island” i “bill_length_mm” del dataframe “penguins”.
# Quina és l'espècie de pingüí més comú a l'illa de Torgersen?

df2 = penguins.loc[:,["species","island","bill_length_mm"]]
df3 = penguins[penguins["island"] == "Torgersen"]
grup = df3.groupby(["species"]).count()

grup
```

```python
#3 Troba tots els pingüins que tenen un “bill_length_mm” més gran de 10.
# Quants pingüins compleixen amb aquesta condició?
x= 3.14
res = penguins[penguins["bill_length_mm"] > 10]
print(f"Hi ha {res.shape[0]} pinguins amb el bec major a 10 mm {x}")
```

```python
#4 Troba tots els pingüins de l'espècie “Gentoo” que tenen un “bill_length_mm” major a 5. Quina és l'illa on es troben?
resultats = penguins[(penguins.species == "Gentoo") &
 (penguins.bill_length_mm > 5)].island

print(resultats.unique()[0])
```

```python
#5 Selecciona les files des de la 20 fins a la 30 al dataframe penguins. Quina és la mitjana de “bill_length_mm” en aquest rang?

files = penguins.loc[20:30,:].bill_length_mm
files.mean()
```
